```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

## Generate random data 生成隨機數據

```python
# 在二維空間中生成一些隨機數據的代碼：總共生成了 100 個數據點並將其分為兩組，每組 50 個點。

X = -2 * np.random.rand(100,2)
X1 = 1 + 2 * np.random.rand(50,2)
X[50:100, :] = X1
plt.scatter(X[ : , 0], X[ :, 1], s = 50, c = "b")
plt.show()
```

## K-Means

```python
from sklearn.cluster import KMeans
Kmean = KMeans(n_clusters=2)  #切分為兩個群
Kmean.fit(X)
```

## Finding the centroid and predict 找到質心並預測

```python
# 切了兩個群，獲得兩組質心位置
Kmean.cluster_centers_
```

```python
plt.scatter(X[ : , 0], X[ : , 1], s =50, c="b")
plt.scatter(-0.94665068, -0.97138368, s=200, c="g", marker="s")
plt.scatter(2.01559419, 2.02597093, s=200, c="r", marker="s")
plt.show()
```

```python
# Here is the code for getting the labels property of the K-means clustering example dataset; 
# that is, how the data points are categorized into the two clusters.
# 這是獲取 K 均值聚類示例數據集的標籤屬性的代碼；數據點如何分類到兩個集群中。

Kmean.labels_   
#得出 50 個數據點屬於 0 集群，而其餘的屬於 1 集群。
```

```python
# predicting the cluster of a data point :
sample_test = np.array([-3.0,-3.0])
print(sample_test)
second_test = sample_test.reshape(1, -1)
print(second_test)
Kmean.predict(second_test)
# It shows that the test data point belongs to the 0 (green centroid) cluster.
# 它表明測試數據點屬於 0（綠色質心）。
```
