```python
### DataSet

CSV_COLUMN_NAMES = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
SPECIES = ['Setosa', 'Versicolor', 'Virginica']

train = pd.read_csv("data/iris_training.csv", names=CSV_COLUMN_NAMES, header=0)
test = pd.read_csv("data/iris_test.csv", names=CSV_COLUMN_NAMES, header=0)

# pop the species column off and use that as our label.
train_y = train.pop('Species')
test_y = test.pop('Species')



### Feature Columns 特色欄目分類

my_feature_columns = []
for key in train.keys():
    my_feature_columns.append(tf.feature_column.numeric_column(key=key))
    
print(my_feature_columns)
```

```python
### Input Function：沒有 epochs，只需要將資料將以 256 個為一組批量

def input_fn(features, labels, training=True, batch_size=256):
    # 將輸入轉換為資料集，使用資料及其標籤建立 tf.data.Dataset 對象
    dataset = tf.data.Dataset.from_tensor_slices((dict(features), labels))

    # 如果您處於訓練模式，則隨機打亂資料並重複
    if training:
        dataset = dataset.shuffle(1000).repeat()

    return dataset.batch(batch_size)



### Building the Model 建立模型：Deep Neural Network

# 2 hidden layers of 30 and 10 nodes respectively (2個隱藏層 30個節點和 10個神經元)並將花朵分為三類
classifier = tf.estimator.DNNClassifier(
    feature_columns=my_feature_columns,
    hidden_units=[30, 10],
    n_classes=3)

# 我們包含一個 lambda 以避免在之前建立內部函數，並且提供 5000 次 epochs
classifier.train(
    input_fn=lambda: input_fn(train, train_y, training=True),
    steps=5000)

# INFO:tensorflow:Loss for final step: 0.2092987.
# 意味著損失為 20
```

```python
# Evaluation 
eval_result = classifier.evaluate(input_fn=lambda: input_fn(test, test_y, training=False))
print('\nTest set accuracy: {accuracy:0.3f}\n'.format(**eval_result))
```

```python
### Predictions 預測

# 當我們預測時 並不知道 labels
def input_fn(features, batch_size=256):
    # Convert the inputs to a Dataset without labels.
    return tf.data.Dataset.from_tensor_slices(dict(features)).batch(batch_size)

features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']
predict = {}

print("Please type numeric values as prompted.")
for feature in features:
    valid = True
    while valid:
        val = input(feature + ": ")
        if not val.isdigit(): valid = False
    predict[feature] = [float(val)]     # 特徵為 ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']

predictions = classifier.predict(input_fn=lambda: input_fn(predict))
for pred_dict in predictions:
    print(pred_dict)
    class_id = pred_dict['class_ids'][0]
    probability = pred_dict['probabilities'][class_id]

    print('Prediction is "{}" ({:.1f}%)'.format(
        SPECIES[class_id], 100 * probability))
    
# 我們需要輸入特徵 ['SepalLength萼片長度'，'SepalWidth萼片寬度'，'PetalLength花瓣長度'，'PetalWidth花瓣寬度']
# [2.4, 2.6, 6.5, 6.3] 
# [5.1, 3.3, 1.7, 0.5] #Setosa
# [5.9, 3.0, 4.2, 1.5] #Versicolor
# [6.9, 3.1, 5.4, 2.1] #Virginica
    
# 分解 print(pred_dict)
# {'logits': array([-4.487072  , -0.04180508,  4.1337194 ], dtype=float32), 
#  'probabilities': array([0.00017755693, 0.015131865, 0.98469061], dtype=float32), 
#  'class_ids': array([2]), 'classes': array([b'2'], dtype=object),
#  'all_class_ids': array([0, 1, 2], dtype=int32), 
#  'all_classes': array([b'0', b'1', b'2'], dtype=object)}
# probabilities 花卉品種機率 array([0.00017755693, 0.015131865, 0.98469061]
# class_ids 花卉品種為 [2] 也就是 Virginica  ＃ SPECIES = ['Setosa', 'Versicolor', 'Virginica']
```
