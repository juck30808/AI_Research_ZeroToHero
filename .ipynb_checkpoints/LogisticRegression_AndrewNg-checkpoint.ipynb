```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('data/marks.txt', header = None)
X = data.iloc[:,:-1]
y = data.iloc[:,2]
data.head()
```

```python
# filter out the applicants that got admitted
admitted = data.loc[y == 1]

# filter out the applicants that din't get admission
not_admitted = data.loc[y == 0]

# plots
plt.scatter(admitted.iloc[:, 0], admitted.iloc[:, 1], s=10, label='Admitted')
plt.scatter(not_admitted.iloc[:, 0], not_admitted.iloc[:, 1], s=10, label='Not Admitted')
plt.legend()
plt.show()
```

## 數據預處理

```python
def featureNormalize(X):
    (m,n) = X.shape
    X_norm = X
    mu = np.zeros(n);
    sigma = np.zeros(n);

    for i in range(n):
        mu[i] = np.mean(X[:,i])
        sigma[i] = np.std(X[:,i])
        X_norm[:,i] = (X_norm[:,i]-mu[i])/sigma[i]
        
    return X_norm
```

```python
X_data = data.iloc[:,0:2].values
y = data.iloc[:,2].values

X_data = featureNormalize(X_data)

ones = np.ones(X_data.shape[0])
X = np.c_[X_data,ones]
```

## 使用梯度下降實現對數機率回歸

```python
# define some functions that will be used to compute the cost.
def sigmoid(X,theta):    
    linear = np.dot(X,theta)         # Computes the weighted sum of inputs
    return 1/(1+np.exp(-linear))     # Activation function used to map any real value between 0 and 1

# define the cost and the gradient function.
def computeCost(X, y, theta):
    J = -np.dot(y.T,np.log(sigmoid(X,theta))) - np.dot((1-y).T,np.log(1-sigmoid(X,theta)))
    return J/len(y);

def gradientDescent(X, y, theta, alpha, num_iters):
    m = len(y)
    J_history = np.zeros(num_iters);
    theta_len = len(theta);

    #迭代 num_iters 次
    for num_iter in range(num_iters):
        theta = theta - (alpha/m)*np.dot(X.T,(sigmoid(X,theta).reshape(-1)-y))
        J_history[num_iter] = computeCost(X, y, theta)
        
    return theta, J_history
```

```python
alpha = 0.01
num_iters = 400
theta = np.zeros(3)

theta, J_history = gradientDescent(X, y, theta, alpha, num_iters)
plt.plot(J_history)
```

```python
y_pred = sigmoid(X,theta)
y_pred[y_pred >= 0.5] = 1
y_pred[y_pred < 0.5] = 0
plt.scatter(X[y_pred==0][:,0], X[y_pred==0][:,1], alpha=.8, color='navy')
plt.scatter(X[y_pred==1][:,0], X[y_pred==1][:,1],alpha=.8, color='turquoise')
```

## Logistic Regression Model Construction 模型建構

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1000.0, random_state=0)
lr.fit(data.iloc[:,:1].values, data.iloc[:,2].values)
```

```python
print('regression Parameters: Intercept',lr.intercept_) #截距
print('regression Parameters: Co-efficient',lr.coef_)   #參數(斜率)
```
